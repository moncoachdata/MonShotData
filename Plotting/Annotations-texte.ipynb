```python
import pandas as pd
import numpy as np

import yfinance
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.rcParams['font.family'] = ['Times New Roman', 'sans-serif']
```

```python
pd.options.plotting.backend = "plotly"
```

```python
raw_data = yfinance.download(tickers = "^GSPC", start = '2020-01-01', 
                              end = '2022-03-31', interval = "1d")
```

### Sans annotations

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(raw_data.index, raw_data.Close)
plt.xlabel("Date", fontweight = "bold", size = 20)
plt.xticks(size = 15, rotation = 45)
plt.ylabel("S&P 500", fontweight = "bold", size = 20)
plt.yticks(size = 15)
plt.title("Graphique simple", fontweight = "bold", size = 30);
```

### Avec annotations

```python
fig, ax = plt.subplots(figsize=(13, 6))

ax.plot(raw_data.index, raw_data.Close)

# Annotation du premier pic
ax.annotate('Première vague', xy=(np.datetime64('2020-02-20'), 3400), xytext=(np.datetime64('2020-01-31'), 4000), size=15, fontweight = "bold", va="center", ha="center",
            arrowprops=dict(arrowstyle="simple", facecolor='orange',
                            connectionstyle="arc3,rad=-0.2"))

# Annotation du second pic
ax.annotate('Deuxième vague', xy=(np.datetime64('2020-09-23'), 3250), xytext=(np.datetime64('2020-11-15'), 2700), size=15, fontweight = "bold", va="center", ha="center",
            arrowprops=dict(arrowstyle="simple", facecolor='orange',
                            connectionstyle="arc3,rad=-0.2"))

# Annotation du troisième pic
ax.annotate('Troisième vague', xy=(np.datetime64('2020-10-30'), 3450), xytext=(np.datetime64('2020-09-15'), 4300), size=15, fontweight = "bold", va="center", ha="center",
            arrowprops=dict(arrowstyle="simple", facecolor='orange',
                            connectionstyle="arc3,rad=-0.2"))

# Annotation du quatrième pic
ax.annotate('Quatrième vague', xy=(np.datetime64('2021-09-30'), 4300), xytext=(np.datetime64('2021-07-31'), 3500), size=15, fontweight = "bold", va="center", ha="center",
            arrowprops=dict(arrowstyle="simple", facecolor='orange',
                            connectionstyle="arc3,rad=0.2"))

# Annotation du cinquième pic
ax.annotate('Cinquième vague', xy=(np.datetime64('2022-01-27'), 4300), xytext=(np.datetime64('2022-02-28'), 3500), size=15, fontweight = "bold", va="center", ha="center",
            arrowprops=dict(arrowstyle="simple", facecolor='orange',
                            connectionstyle="arc3,rad=-0.2"))

plt.xlabel("Date", fontweight = "bold", size = 20)
plt.xticks(size = 15, rotation = 45)
plt.ylabel("S&P 500", fontweight = "bold", size = 20)
plt.yticks(size = 15)
plt.title("Graphique Annotée", fontweight = "bold", size = 30);
```
