```python
import numpy as np
import perfplot

perfplot.plot(
    setup = lambda n: np.random.rand(n),  # ou setup = np.random.rand
    kernels = [
        lambda a: np.c_[a, a],
        lambda a: np.stack([a, a]).T,
        lambda a: np.vstack([a, a]).T,
        lambda a: np.column_stack([a, a])
    ],
    labels = ["c_", "stack", "vstack", "col_stack"],
    n_range = [2**k for k in range(5,20)]
)
```
