# Utilise les diagrammes en boîte avec prudence ! Ils peuvent induire en erreur.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

plt.rcParams['figure.dpi'] = 300
```

```python
s1_pre = np.random.normal(0, 1, size=5000)

s1 = [i for i in s1_pre if i<2.6 and i>-2.6]
```

```python
# Créer de nouveaux datasets

s2 = []
for i in s1:
    if i>0.05 and i<0.6:
        s2.append(0.65)
    elif i<-0.05 and i>-0.6:
        s2.append(-0.65)
    else:
        s2.append(i)

s3 = []
for i in s1:
    if i>-0.4 and i<0.4:
        s3.append(0)
    else:
        s3.append(i)
```

```python
plt.style.use(["seaborn-v0_8-darkgrid"])
```

```python
c_list = ['#46C7C7', '#00B9FF', '#FF7051'][::-1]

n = 25
f, axes = plt.subplots(3, 3 ,figsize=(15,9))
ax1=sns.boxplot(s1,ax= axes[0,0],color= c_list[0],boxprops= dict(alpha=.9))
ax2=sns.boxplot(s2,ax= axes[0,1],color= c_list[1],boxprops= dict(alpha=.9))
ax3=sns.boxplot(s3,ax= axes[0,2],color= c_list[2],boxprops= dict(alpha=.9))

ax4=sns.histplot(s1, bins=n, ax=axes[1,0], color=c_list[0])
ax5=sns.histplot(s2, bins=n, ax=axes[1,1], color=c_list[1])
ax6=sns.histplot(s3, bins=n, ax=axes[1,2], color=c_list[2])

ax7=sns.violinplot(s1, ax=axes[2,0], color=c_list[0])
ax8=sns.violinplot(s2, ax=axes[2,1], color=c_list[1])
ax9=sns.violinplot(s3, ax=axes[2,2], color=c_list[2])

ax4.set(ylabel=None, yticklabels=[])
ax5.set(ylabel=None, yticklabels=[])
ax6.set(ylabel=None, yticklabels=[])

ax1.set_title("Dataset A", fontsize = 15, fontweight="bold")
ax2.set_title("Dataset B", fontsize = 15, fontweight="bold")
ax3.set_title("Dataset C", fontsize = 15, fontweight="bold")

ax1.set_ylabel("Diagramme en boîte", fontsize = 15)
ax4.set_ylabel("Histogramme", fontsize = 15)
ax7.set_ylabel("Diagramme en violon", fontsize = 15)

plt.savefig("final_boxes.jpeg", dpi = 600, bbox_inches="tight")
plt.show()
```
