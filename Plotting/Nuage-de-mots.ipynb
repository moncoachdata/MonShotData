```python
from wordcloud import WordCloud, ImageColorGenerator
import PIL.Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
text = """Python est un langage de programmation général de haut niveau. Sa philosophie de conception met l'accent sur la lisibilité du code avec l'utilisation d'une indentation significative.
Python est à typage dynamique et à collecte de déchets. Il prend en charge de nombreux paradigmes de programmation, notamment la programmation structurée (en particulier procédurale), orientée objet et fonctionnelle. Il est souvent décrit comme un langage « à piles incluses » en raison de sa bibliothèque standard complète.
Guido van Rossum a commencé à travailler sur Python à la fin des années 1980 comme successeur du langage de programmation ABC et l'a publié pour la première fois en 1991 sous le nom de Python 0.9.0. Python 2.0 a été publié en 2000 et a introduit de nouvelles fonctionnalités telles que la compréhension des listes, la collecte des déchets par détection de cycle, le comptage des références et la prise en charge de l'Unicode. Python 3.0, publié en 2008, est une révision majeure qui n'est pas totalement rétrocompatible avec les versions antérieures. Python 2.7.18, publié en 2020, est la dernière version de Python 2.
Sa philosophie de base est résumée dans le document The Zen of Python (PEP 20), qui comprend des aphorismes tels que :
Le beau est mieux que le laid.
L'explicite vaut mieux que l'implicite.
La simplicité est préférable à la complexité.
Mieux vaut complexe que compliqué.
La lisibilité est importante.
Python se veut un langage facilement lisible. Sa mise en forme est visuellement épurée et utilise souvent des mots-clés anglais là où d'autres langages utilisent la ponctuation. Contrairement à de nombreux autres langages, il n'utilise pas de parenthèses pour délimiter les blocs, et les points-virgules après les déclarations sont autorisés mais rarement utilisés. Il comporte moins d'exceptions syntaxiques et de cas particuliers que le C ou le Pascal."""
text = text.replace("\n", " ")
```

```python
text
```

```python
wc = WordCloud(min_word_length=5).generate(text)
plt.axis('off')
plt.imshow(wc)
plt.show()
```

```python
msk = np.array(PIL.Image.open("py-logo.jpeg"))
colors = ImageColorGenerator(msk)
wc = WordCloud(mask=msk, min_word_length=5,
               background_color="white",
               min_font_size=10, repeat = True).generate(text)
wc.recolor(color_func = colors)
plt.axis('off')
plt.imshow(wc)
plt.show()
```
