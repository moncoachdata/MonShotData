# Tracé de régression linéaire polynomiale facilité par Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = 35
x = np.linspace(3, 9, n)
y = 0.8*x**2 - 10*x + 3 + 1*np.random.random(n)

df = pd.DataFrame()
df['x'] = x
df['y'] = y
```

```python
df.head()
```

```python
sns.lmplot(x = "x", 
           y = "y", 
           data = df, 
           order = 2);
```
