# Diagrammes de dispersion trop chargés ? Voici 3 façons simples de les épurer.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

sns.set()
plt.style.use(["seaborn-v0_8-darkgrid"])
```

```python
df=pd.DataFrame({'x': np.random.normal(10, 1.2, 20000), 'y': np.random.normal(10, 1.2, 20000), 'group': np.repeat('A',20000) })
tmp1=pd.DataFrame({'x': np.random.normal(14.5, 1.2, 20000), 'y': np.random.normal(14.5, 1.2, 20000), 'group': np.repeat('B',20000) })
tmp2=pd.DataFrame({'x': np.random.normal(9.5, 1.5, 20000), 'y': np.random.normal(15.5, 1.5, 20000), 'group': np.repeat('C',20000) })
df=pd.concat((df, tmp1,tmp2))
df = df.reset_index(drop = True)

df.head()
```

```python
X = df[["x", "y"]].values
```

## Diagramme de dispersion par défaut

```python
fig, ax = plt.subplots(figsize = (12, 9))
plt.scatter(X[:, 0], X[:, 1])
ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.tick_params(axis='both', which='major', pad=15)
ax.set_title("Diagramme de dispersion par défaut", fontsize = 25, fontweight="bold")
plt.xlim(3, 20);
```

## Diagramme de dispersion avec réduction de la taille des marqueurs

```python
fig, ax = plt.subplots(figsize = (12, 9))
plt.scatter(X[:, 0], X[:, 1], s=0.9)
ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.tick_params(axis='both', which='major', pad=15)
ax.set_title("Diagramme de dispersion\navec réduction de la taille des marqueurs", fontsize = 25, fontweight="bold")
plt.xlim(3, 20);
```

## Diagrammes hexagonaux

```python
ax = df.plot(kind = "hexbin",
        x = 'x',
        y = 'y', gridsize = 13, figsize=(12, 9),  sharex=False, cmap="Blues")

ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.set_title("Diagramme hexbin", fontsize = 25, fontweight="bold")
ax.tick_params(axis='both', which='major', pad=15);
```

## Diagramme KDE

```python
fig, ax = plt.subplots(figsize = (12, 9))
sns.kdeplot(data = df, x="x", y="y", cmap="Blues", shade=True)
plt.xlim(4, 20)
plt.ylim(5, 22)
ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.set_title("Diagramme KDE", fontsize = 25, fontweight="bold")
ax.tick_params(axis='both', which='major', pad=15);
```
