# Une astuce simple qui améliore considérablement la qualité des tracés de Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
sns.set()
import matplotlib.ticker as mtick

plt.style.use("seaborn-v0_8-darkgrid")
```

```python
def percent_formatter(x, pos):
    return f"{x*100:.0f}%"
```

```python
# Définir la graine (seed) aléatoire pour la reproductibilité
np.random.seed(42)

# Nombre d'échantillons
n_samples = 10000

# Générer des probabilités selon la loi de Benford
digits = np.arange(1, 10)
probabilities = np.log10(1 + 1/digits)

# Générer les premiers chiffres en utilisant les probabilités
first_digits = np.random.choice(digits, size=n_samples, p=probabilities)

# Calculer la fréquence de chaque premier chiffre
frequency, _ = np.histogram(first_digits, bins=np.arange(1, 11))
frequency = frequency/np.sum(frequency)
```

## Graphique par défaut

```python
fig, ax = plt.subplots()

# Créer un diagramme à barres
plt.bar(digits, frequency, color = '#46C7C7')
plt.xlabel('Premier chiffre')
plt.ylabel('Fréquence')
plt.xticks(range(1, 10))
ax.yaxis.set_major_formatter(mtick.FuncFormatter(percent_formatter))

plt.title("Loi de Benford - Distribution des premiers chiffres")
plt.show()
```
