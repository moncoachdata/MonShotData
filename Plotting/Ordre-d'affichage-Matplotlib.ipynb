# Matplotlib possède de nombreux secrets. En voici un !

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.style.use(["seaborn-v0_8-darkgrid"])
c_list = ['#46C7C7', '#00B9FF', '#FF7051'][::-1]
sns.set()
```

## Données

```python
r = np.linspace(0.3, 1, 30)
theta = np.linspace(0, 4*np.pi, 30)
x = r * np.sin(theta)
y = r * np.cos(theta)
```

## Disposition par défaut

```python
fig, ax = plt.subplots(1, figsize=(5, 5))

ax.plot(x, y, lw=5, color = c_list[0])
ax.scatter(x, y, s=200  , color = c_list[1])
ax.set_title('Lignes au dessus', fontsize = 20, fontweight="bold")
ax.set_yticklabels([])
ax.set_xticklabels([]);
```

## Contrôle de l'ordre d'affichage

```python
fig, ax = plt.subplots(1, figsize=(6, 6))

ax.plot(x, y, lw=5,c = c_list[0],  zorder = 1)
ax.scatter(x, y, s=200,  c = c_list[1], zorder = 2)  # déplacer les points sur le dessus de la ligne
ax.set_title('Points au dessus', fontsize = 20, fontweight="bold")
ax.set_yticklabels([])
ax.set_xticklabels([]);
```

## Lignes en dessous de la grille

```python
fig, ax = plt.subplots(1, figsize=(6, 6))

ax.plot(x, y, lw=7,c = c_list[0],  zorder = 0)
ax.scatter(x, y, s=200,  c = c_list[1], zorder = 2)  # déplacer les points au-dessus des lignes
ax.set_title('Lignes en dessous de la grille', fontsize = 20, fontweight="bold")
ax.set_yticklabels([])
ax.set_xticklabels([]);
```
