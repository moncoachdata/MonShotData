# Les diagrammes Ã  hexagones, une alternative plus riche aux diagrammes de dispersion

```python
import pandas as pd
```

```python
df = pd.read_csv("../_datasets/exemple_diagramme_hexbin.csv")
```

## Scatter

```python
df.plot(kind = 'scatter', x = 'x' , y = 'y');
```

## Hexbin

```python
df.plot(kind = 'hexbin',
        x = 'x',
        y = 'y', 
        gridsize = 13);
```
