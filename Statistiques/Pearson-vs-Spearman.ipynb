# La limite de la corrélation de Pearson que beaucoup ignorent souvent

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr
```

```python
plt.rcParams['font.family'] = ['Times New Roman', 'sans-serif']
sns.set()
```

### Données

```python
n = 200
x = np.linspace(-6, 6, n)


linear_data = 2*x+3

# Générer des données polynomiales
poly_data = x**3 + x**2 + 2*x +2
 
# Ajouter du bruit aux deux ensembles de données
linear_data = linear_data + np.random.normal(0, 2, size=linear_data.shape)
poly_data = poly_data + np.random.normal(0, 7, size=poly_data.shape)
poly_data_neg = -poly_data
```

```python
ax_ = sns.scatterplot(x=x, y=poly_data, s = 70)
ax_.set(xlabel=None, ylabel=None, yticklabels=[], xticklabels=[])
ax_.set_title("Données non linéaires", size = 20, fontweight = "bold")
plt.show();
```

```python
ax_ = sns.scatterplot(x=x, y=linear_data, s = 70)
ax_.set(xlabel=None, ylabel=None, yticklabels=[], xticklabels=[])
ax_.set_title("Données linéaires", size = 20, fontweight = "bold")
plt.show();
```

```python
# Calculer le coefficient de corrélation de Pearson pour des données linéaires
linear_pearson, _ = pearsonr(x, linear_data)
print("Corrélation de Pearson pour les données linéaires :", linear_pearson)

# Calculer le coefficient de corrélation de Spearman pour les données linéaires
linear_spearman, _ = spearmanr(x, linear_data)
print("Corrélation de Spearman pour les données linéaires :", linear_spearman)

# Calculer le coefficient de corrélation de Pearson pour des données polynomiales
poly_pearson, _ = pearsonr(x, poly_data)
print("Corrélation de Pearson pour les données polynomiales :", poly_pearson)

# Calculer le coefficient de corrélation de Spearman pour des données polynomiales
poly_spearman, _ = spearmanr(x, poly_data)
print("Corrélation de Spearman pour les données polynomiales :", poly_spearman)
```
