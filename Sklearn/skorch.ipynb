```python
import torch
import numpy as np
from skorch import NeuralNetClassifier
```

```python
# G√©n√©rer des donn√©es synth√©tiques
X = np.random.randn(10, 3).astype(np.float32)
y = np.random.randint(0, 2, (10,)).astype(np.float32)

X = torch.from_numpy(X)   # convertir en tenseurs
y = torch.from_numpy(y).type(torch.LongTensor)
```

```python
# D√©finir un mod√®le PyTorch simple
class MyModel(torch.nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.fc1 = torch.nn.Linear(3, 20)
        self.fc2 = torch.nn.Linear(20, 2)
    
    def forward(self, x):
        x = self.fc1(x)
        x = torch.nn.functional.relu(x)
        x = self.fc2(x)
        return x
```

```python
# Cr√©er un NeuralNetClassifier √† l'aide de skorch
model = NeuralNetClassifier(
    MyModel,
    max_epochs=10,
    lr=0.1,
    criterion=torch.nn.CrossEntropyLoss
)
```

```python
# Utiliser l'API scikit-learn pour entra√Æner et pr√©dire
model.fit(X, y)
predictions = model.predict(X)
```
