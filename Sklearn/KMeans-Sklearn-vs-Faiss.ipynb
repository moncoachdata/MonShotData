```python
import pandas as pd
import faiss
from sklearn.cluster import KMeans
```

## Charger les donn√©es

```python
data = pd.read_csv("mnist_train.csv")
```

```python
data = data.dropna()
```

```python
data
```

```python
x = data.drop(columns = ["label"]).values
y = data["label"].values
```

## KMeans de Sklearn

```python
%%timeit -n1 -r5

kmeans = KMeans(n_init='auto', n_clusters=8).fit(x)
```

## KMeans de Faiss

```python
%%timeit -n1 -r5

kmeans = faiss.Kmeans(d=784, k=8)
kmeans.train(x)
```
