# Prévisualiser un arbre de décision à l'aide d'une ligne Sklearn

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
X = df.iloc[:, [0,1,2,3]].values
y = df.iloc[:,[-1]].values
```

```python
df["species"] = pd.factorize(df["species"])[0]
```

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
classes = ['setosa', 'versicolor', 'virginica']
```

```python
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree, export_text
```

```python
my_tree = DecisionTreeClassifier(max_depth=2)
my_tree.fit(X, y)
```

## Méthode 1

```python
plot_tree(my_tree, feature_names=features, class_names=classes, filled=True);
```

## Méthode 2

```python
print(export_text(my_tree, feature_names=features))
```
