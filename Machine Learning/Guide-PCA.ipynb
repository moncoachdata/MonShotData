```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
sns.set()
```

```python
x = np.linspace(1, 5, 100)
y = 2 * x + 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ensemble de données linéaires avec du bruit : y = 2x + 3')
plt.show()
```

```python
data = np.column_stack((x, y))
```

```python
pca = PCA(n_components=1)
pca.fit(data)
```

```python
# Projeter les données sur la première composante principale
projected_data = pca.transform(data)

# Reconstruire les données à l'aide de la première composante principale
reconstructed_data = pca.inverse_transform(projected_data)

plt.scatter(x, y, label="Données d'origine")
plt.scatter(reconstructed_data[:, 0], reconstructed_data[:, 1], label='Données reconstruites', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('PCA à 1 composante')
plt.legend()
plt.show()
```
