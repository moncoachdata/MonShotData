# Clustering non supervisé : les 3 métriques essentielles à connaitre

[Lien du post MonShotData](https://www.monshotdata.com/p/performance-clustering-non-supervise)

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

import seaborn as sns
sns.set()
```

```python
# Générer un ensemble de données de blobs 2D avec trois centroïdes
X, y_true = make_blobs(n_samples=300, centers=4, n_features=2)

plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis')
plt.title('Clustering K-Means')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
colors = ['#fe7c73', '#2471A3', '#3498DB', '#27AE60', '#82E0AA', '#D35400', '#5D6D7E', '#E74C3C', '#21618C', '#B7950B', '#46C7C7', '#00B9FF', '#FF7051', "orange", "darkorange", "tomato", "coral", "limegreen", "lightsalmon"]
```

```python
kmeans = KMeans(n_clusters=4)
y_pred = kmeans.fit_predict(X)

filtered = [colors[13], colors[15], colors[2], colors[3]]
fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=[filtered[i] for i in y_pred], s = 70)
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 90, c="black", marker='o', label='Centroïdes')
ax.set_title('K-Means (4 centroïdes)', fontsize = 20, weight="bold", pad = 10)

ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()


# Calculer le coefficient de silhouette
silhouette_coefficient = silhouette_score(X, y_pred)

# Calcul de l'indice de Calinski-Harabasz
calinski_harabasz_index = calinski_harabasz_score(X, y_pred)

# Calcul de l'indice de Davies-Bouldin
davies_bouldin_index = davies_bouldin_score(X, y_pred)

# Imprimer les metrics d'évaluation
print("Coefficient de silhouette :", silhouette_coefficient)
print("Indice Calinski-Harabasz :", calinski_harabasz_index)
print("Indice Davies-Bouldin :", davies_bouldin_index)
```

```python
kmeans = KMeans(n_clusters=6, init="random")
y_pred = kmeans.fit_predict(X)

filtered = [colors[13], colors[15], colors[2], colors[3], colors[10], colors[5]]
fig, ax = plt.subplots()

ax.scatter(X[:, 0], X[:, 1], c=[filtered[i] for i in y_pred], s = 70)
ax.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 90, c="black", marker='o', label='Centroïdes')
ax.set_title('K-Means (6 centroïdes)', fontsize = 20, weight="bold", pad = 10)

ax.set_xticklabels([])
ax.set_yticklabels([])

plt.show()


# Calculer le coefficient de silhouette
silhouette_coefficient = silhouette_score(X, y_pred)

# Calcul de l'indice de Calinski-Harabasz
calinski_harabasz_index = calinski_harabasz_score(X, y_pred)

# Calcul de l'indice de Davies-Bouldin
davies_bouldin_index = davies_bouldin_score(X, y_pred)

# Imprimer les metrics d'évaluation
print("Coefficient de silhouette :", silhouette_coefficient)
print("Indice Calinski-Harabasz :", calinski_harabasz_index)
print("Indice Davies-Bouldin :", davies_bouldin_index)
```
