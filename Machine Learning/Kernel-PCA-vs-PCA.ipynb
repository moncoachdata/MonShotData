# Au-delà du PCA : que faire quand tes données ne sont pas linéairement séparables ?

[Lien du post MonShotData](https://www.monshotdata.com/p/pca-que-faire-quand-tes-donnees-ne-sont-pas-lineairement-separables)

```python
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.decomposition import KernelPCA, PCA

sns.set()
```

```python
colors = ['#fe7c73', '#2471A3']
```

## Dataset

```python
X, y = make_moons(n_samples=500, noise=0.03)

fig, ax = plt.subplots()

plt.scatter(X[:, 0], X[:, 1], c=[colors[i] for i in y])
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

## PCA

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

fig, ax = plt.subplots()

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=[colors[i] for i in y])

ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

## Kernel PCA

```python
kpca = KernelPCA(n_components=2, kernel='rbf', gamma=14)
X_kpca = kpca.fit_transform(X)

fig, ax = plt.subplots()

plt.scatter(X_kpca[:, 0], X_kpca[:, 1], c=[colors[i] for i in y])
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```
