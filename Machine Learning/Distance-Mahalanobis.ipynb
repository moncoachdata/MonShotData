# La distance euclidienne n'est pas infaillible : voilà pourquoi

```python
import numpy as np
from scipy.spatial.distance import mahalanobis
import matplotlib.pyplot as plt
```

```python
# Générer les données linéaires avec du bruit ajouté
n = 50
x = np.linspace(1, 5, n)
y = 2 * x + 3 + np.random.normal(size=n)

# Tracer les données
plt.plot(x, y, 'o', label='data')

# Ajouter des étiquettes d'axe et une légende
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()

# Afficher le graphique
plt.show()
```

```python
# Calculer la moyenne et la matrice de covariance des données
data = np.column_stack((x, y))
mean = np.mean(data, axis=0)
cov = np.cov(data.T)
```

```python
p1 = np.array([5, 13])
mahalanobis(p1, mean, np.linalg.inv(cov))
```

```python
p2 = np.array([0, 11])   ## la droite reliant p2 et la moyenne est perpendiculaire à la droite reliant p1 et la moyenne 
mahalanobis(np.array(p2), mean, np.linalg.inv(cov))
```
