# Régression linéaire sous-performante ? Voici comment résoudre le problème

[Lien du post MonShotData](https://www.monshotdata.com/p/asymetrie-regression-lineaire)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

import seaborn as sns
sns.set()
```

```python
color_code = '#fe7c73'
```

```python
data = np.random.gamma(2, 1, 1000) * 10
```

```python
plt.hist(data, bins=30, edgecolor='black', color = color_code)
plt.title('Distribution initiale des données', fontsize = 20, fontweight = "bold");
```

```python
# Appliquer une transformation logarithmique
data_log = np.log(data)

# Tracer la distribution des données log-transformées
plt.hist(data_log, bins=30, edgecolor='black', color = color_code)
plt.title('Données transformées par logarithme', fontsize = 20, fontweight = "bold");
```

```python
# Appliquer la transformation de la racine carrée
data_sqrt = np.sqrt(data)

# Tracer la distribution des données transformées par la racine carrée
plt.hist(data_sqrt, bins=30, edgecolor='black', color = color_code)
plt.title('Données transformées par racine carrée', fontsize = 20, fontweight = "bold");
```

```python
# Appliquer la transformation de Box-Cox
data_boxcox, _ = stats.boxcox(data)

# Tracer la distribution des données transformées selon la transformation de Box-Cox
plt.hist(data_boxcox, bins=30, edgecolor='black', color = color_code)
plt.title('Données transformées par Box-Cox', fontsize = 20, fontweight = "bold");
```
